```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y= [10,20,30,40,50,60]
```

```python
plt.xlabel("Days")#it labels the x axis
plt.ylabel("Stocks")# it labels the y axis
plt.plot(x,y,marker="o",linestyle=":",color="purple",ms=13,mec="black",mfc="white") # it plots the graph
plt.title("Sales") # gives a title to the graph
plt.show() # it displays the graph
```

```python
import numpy as np

x = np.array([1,4,5,6,8,3,1])
y = np.array([2,5,7,2,2,3,4])
z= np.array([3,2,5,4,5,6,7,2,3,5,8])

plt.plot(x,label="x")
plt.plot(y,label="y")
plt.plot(z,label="z")

plt.grid(axis="y")
plt.show()
```
