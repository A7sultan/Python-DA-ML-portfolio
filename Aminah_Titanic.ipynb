Dropping columns with non-numeric and unimportant data

```python
cols_to_drop=['Name','Ticket','Cabin']

tita =tita.drop(cols_to_drop,axis = 1)
tita.info()
```

missing values-

There are still missings values in age,embarked and there are object values in sex and embarked
embarked and sex are imp colms so we cant drop them.


so we convert these colms to individual colms whic have numerical data in the form of 0's and 1's
we create a dummy variable.

```python
dummies =[]
cols = ['Sex','Embarked']
for col in cols:
  dummies.append(pd.get_dummies(tita[col]))
```

```python
tita_dummies = pd.concat(dummies,axis=1)   #concatenate the dummies
```

```python
tita_dummies # here the sex values male and female are turned into colmns and they have numerical data
            # same for embarked as well, there are seperate colms for C,Q and S
```

```python
# now concate these new colms to the orignal df
tita = pd.concat((tita,tita_dummies),axis =1 )
tita
```

Dropping duplicate columns

```python
tita = tita.drop(['Sex','Embarked'],axis=1)
tita
```

```python
tita['Age'].interpolate(inplace=True)  #age is important but is missng some data
tita.info()
```
